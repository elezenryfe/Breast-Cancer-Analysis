```python
%matplotlib inline
import matplotlib
import random as rd
import numpy as np
import pandas as pd
import math as math
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"data - Copy.csv",header = None)
x = df.copy()
del x[0]
del x[1]
```

```python
x = (x-x.mean())/(x.std())
x = np.array(x)
store = []
m = x.shape[0]
n = x.shape[1]
iteration = 80
k = 5
centroids = np.array([]).reshape(n,0)
first = []
second = []
third = []
fourth = []
fifth = []
for i in range(k):
    rand = rd.randint(0,m-1)
    centroids = np.c_[centroids,x[rand]]
    store.append(rand)
Result = {}
for i in range(iteration):
    Euc_dist=np.array([]).reshape(m,0)
    for j in range(k):
        tempDist=np.sum((x-centroids[:,j])**2,axis=1)
        Euc_dist=np.c_[Euc_dist,tempDist]
    C=np.argmin(Euc_dist,axis=1)+1

    Y={}
    for j in range(k):
        Y[j+1]=np.array([]).reshape(n,0)
    for i in range(m):
        Y[C[i]]=np.c_[Y[C[i]],x[i]]     
    for l in range(k):
        Y[l+1]=Y[l+1].T    
    for u in range(k):
        centroids[:,u]=np.mean(Y[u+1],axis=0)
    Result=Y
for i in range(len(C)):
    if C[i] == 1:
        first.append(x[i])
    if C[i] == 2:
        second.append(x[i])
    if C[i] == 3:
        third.append(x[i])
    if C[i] == 4:
        fourth.append(x[i])
    if C[i] == 5:
        fifth.append(x[i])
cluster1 = []
cluster2 = []
cluster3 = []
cluster4 = []
cluster5 = []
first = np.array(first)
second = np.array(second)
third = np.array(third)
fourth = np.array(fourth)
fifth = np.array(fifth)

radius2 = np.round(math.sqrt(np.round(np.sum((second-centroids[:,1])**2),3)/len(second)),3)
radius1 = np.round(math.sqrt(np.round(np.sum((first-centroids[:,0])**2),3)/len(first)),3)
```

```python
if k > 2:
    radius3 = np.round(math.sqrt(np.round(np.sum((third-centroids[:,2])**2),3)/len(third)),3)
    if k == 5:
        radius4 = np.round(math.sqrt(np.round(np.sum((fourth-centroids[:,3])**2),3)/len(fourth)),3)
        radius5 = np.round(math.sqrt(np.round(np.sum((fifth-centroids[:,4])**2),3)/len(fifth)),3)
dij12 = np.round(math.sqrt(np.sum((centroids[:,0]-centroids[:,1])**2)),3)
cluster1.append((radius1+radius2)/dij12)
cluster2.append((radius2+radius1)/dij12)

if k > 2:
    dij13 = np.round(math.sqrt(np.sum((centroids[:,0]-centroids[:,2])**2)),3)
    cluster1.append((radius1+radius3)/dij13)
    cluster3.append((radius3+radius1)/dij13)
    dij23 = np.round(math.sqrt(np.sum((centroids[:,1]-centroids[:,2])**2)),3)
    cluster2.append((radius2+radius3)/dij23)
    cluster3.append((radius3+radius2)/dij23)
    if k == 5:
        dij14 = np.round(math.sqrt(np.sum((centroids[:,0]-centroids[:,3])**2)),3)
        cluster1.append((radius1+radius4)/dij14)
        cluster4.append((radius4+radius1)/dij14)
        dij24 = np.round(math.sqrt(np.sum((centroids[:,1]-centroids[:,3])**2)),3)
        cluster2.append((radius2+radius4)/dij24)
        cluster4.append((radius4+radius2)/dij24)
        dij34 = np.round(math.sqrt(np.sum((centroids[:,2]-centroids[:,3])**2)),3)
        cluster3.append((radius3+radius4)/dij34)
        cluster4.append((radius4+radius3)/dij34)
        dij15 = np.round(math.sqrt(np.sum((centroids[:,0]-centroids[:,4])**2)),3)
        cluster1.append((radius1+radius5)/dij15)
        cluster5.append((radius5+radius1)/dij15)
        dij25 = np.round(math.sqrt(np.sum((centroids[:,1]-centroids[:,4])**2)),3)
        cluster2.append((radius2+radius5)/dij25)
        cluster5.append((radius5+radius2)/dij25)
        dij35 = np.round(math.sqrt(np.sum((centroids[:,2]-centroids[:,4])**2)),3)
        cluster3.append((radius3+radius5)/dij35)
        cluster5.append((radius5+radius3)/dij35)
        dij45 = np.round(np.sum((centroids[:,3]-centroids[:,4])**2),3)
        cluster4.append((radius4+radius5)/dij45)
        cluster5.append((radius5+radius4)/dij45)

if k == 2:
    DB = (max(cluster1)+max(cluster2))/k
if k == 3:
    DB = (max(cluster1)+max(cluster2)+max(cluster3))/k
if k == 5:
    DB = (max(cluster1)+max(cluster2)+max(cluster3)+max(cluster4)+max(cluster5))/k
print(DB)
#print(Result)
#print(store)
#print(centroid)
```
